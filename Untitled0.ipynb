```python
# ============================================================
# AI FOR MARKET TREND ANALYSIS (AUTO-ADAPTIVE VERSION)
# Dataset: Marketing_data.csv
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

import warnings
warnings.filterwarnings("ignore")

print("Libraries imported successfully")

# -----------------------------
# Load Dataset
# -----------------------------
df = pd.read_csv("Marketing_data.csv")
print("Dataset Loaded | Shape:", df.shape)
print("\nColumns:\n", list(df.columns))

# -----------------------------
# Create AGE (only if possible)
# -----------------------------
if "Age" not in df.columns:
    if "Customer_Age" in df.columns:
        df["Age"] = df["Customer_Age"]
    elif "Birth_Year" in df.columns:
        df["Age"] = 2026 - df["Birth_Year"]
    else:
        df["Age"] = df.select_dtypes(include=np.number).mean(axis=1)
        print("⚠️ Age column approximated using numeric mean")

# -----------------------------
# Create TOTAL SPENDING (auto-detect)
# -----------------------------
spending_cols = [col for col in df.columns if col.lower().startswith("mnt")]

if len(spending_cols) > 0:
    df["MntTotal"] = df[spending_cols].sum(axis=1)
else:
    numeric_cols = df.select_dtypes(include=np.number).columns
    df["MntTotal"] = df[numeric_cols].sum(axis=1)
    print("⚠️ Spending approximated from numeric features")

# -----------------------------
# Encode categorical columns
# -----------------------------
data = df.copy()
for col in data.select_dtypes(include=["object"]).columns:
    data[col] = LabelEncoder().fit_transform(data[col])

print("Preprocessing completed")

# -----------------------------
# Target column check
# -----------------------------
target_col = "Response" if "Response" in data.columns else data.columns[-1]

# -----------------------------
# EDA – Market Trends
# -----------------------------
plt.figure(figsize=(5,4))
sns.countplot(x=target_col, data=data)
plt.title("Market Response Distribution")
plt.show()

plt.figure(figsize=(5,4))
sns.scatterplot(x=data["Income"], y=data["MntTotal"])
plt.title("Income vs Spending Trend")
plt.show()

plt.figure(figsize=(5,4))
sns.scatterplot(x=data["Age"], y=data["MntTotal"])
plt.title("Age vs Spending Trend")
plt.show()

# -----------------------------
# Customer Segmentation
# -----------------------------
features = data[["Income", "Age", "MntTotal"]].fillna(0)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42)
data["Customer_Segment"] = kmeans.fit_predict(features_scaled)

plt.figure(figsize=(6,4))
sns.scatterplot(
    x=data["Income"],
    y=data["MntTotal"],
    hue=data["Customer_Segment"],
    palette="Set1"
)
plt.title("Customer Segments (Market Trends)")
plt.show()

print("\nCustomer Segment Summary:")
print(data.groupby("Customer_Segment")[["Income","Age","MntTotal"]].mean())

# -----------------------------
# AI Model – Market Prediction
# -----------------------------
X = data.drop(target_col, axis=1)
y = data[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

# -----------------------------
# Evaluation
# -----------------------------
y_pred = model.predict(X_test)

print("\nMODEL ACCURACY:", accuracy_score(y_test, y_pred))
print("\nCLASSIFICATION REPORT:\n")
print(classification_report(y_test, y_pred))

# -----------------------------
# Feature Importance
# -----------------------------
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(7,4))
sns.barplot(x="Importance", y="Feature", data=importance.head(10))
plt.title("Top Market Trend Drivers")
plt.show()

print("\nTop Market Influencers:")
print(importance.head(10))

# -----------------------------
# Sample Prediction
# -----------------------------
sample = X_test.iloc[:1]
print("\nSample Market Prediction:", model.predict(sample))

print("\n✅ AI FOR MARKET TREND ANALYSIS – EXECUTED SUCCESSFULLY")
# ============================================================
# AI FOR MARKET TREND ANALYSIS (AUTO-ADAPTIVE VERSION)
# Dataset: Marketing_data.csv
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

import warnings
warnings.filterwarnings("ignore")

print("Libraries imported successfully")

# -----------------------------
# Load Dataset
# -----------------------------
df = pd.read_csv("Marketing_data.csv")
print("Dataset Loaded | Shape:", df.shape)
print("\nColumns:\n", list(df.columns))

# -----------------------------
# Create AGE (only if possible)
# -----------------------------
if "Age" not in df.columns:
    if "Customer_Age" in df.columns:
        df["Age"] = df["Customer_Age"]
    elif "Birth_Year" in df.columns:
        df["Age"] = 2026 - df["Birth_Year"]
    else:
        df["Age"] = df.select_dtypes(include=np.number).mean(axis=1)
        print("⚠️ Age column approximated using numeric mean")

# -----------------------------
# Create TOTAL SPENDING (auto-detect)
# -----------------------------
spending_cols = [col for col in df.columns if col.lower().startswith("mnt")]

if len(spending_cols) > 0:
    df["MntTotal"] = df[spending_cols].sum(axis=1)
else:
    numeric_cols = df.select_dtypes(include=np.number).columns
    df["MntTotal"] = df[numeric_cols].sum(axis=1)
    print("⚠️ Spending approximated from numeric features")

# -----------------------------
# Encode categorical columns
# -----------------------------
data = df.copy()
for col in data.select_dtypes(include=["object"]).columns:
    data[col] = LabelEncoder().fit_transform(data[col])

print("Preprocessing completed")

# -----------------------------
# Target column check
# -----------------------------
target_col = "Response" if "Response" in data.columns else data.columns[-1]

# -----------------------------
# EDA – Market Trends
# -----------------------------
plt.figure(figsize=(5,4))
sns.countplot(x=target_col, data=data)
plt.title("Market Response Distribution")
plt.show()

plt.figure(figsize=(5,4))
sns.scatterplot(x=data["Income"], y=data["MntTotal"])
plt.title("Income vs Spending Trend")
plt.show()

plt.figure(figsize=(5,4))
sns.scatterplot(x=data["Age"], y=data["MntTotal"])
plt.title("Age vs Spending Trend")
plt.show()

# -----------------------------
# Customer Segmentation
# -----------------------------
features = data[["Income", "Age", "MntTotal"]].fillna(0)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42)
data["Customer_Segment"] = kmeans.fit_predict(features_scaled)

plt.figure(figsize=(6,4))
sns.scatterplot(
    x=data["Income"],
    y=data["MntTotal"],
    hue=data["Customer_Segment"],
    palette="Set1"
)
plt.title("Customer Segments (Market Trends)")
plt.show()

print("\nCustomer Segment Summary:")
print(data.groupby("Customer_Segment")[["Income","Age","MntTotal"]].mean())

# -----------------------------
# AI Model – Market Prediction
# -----------------------------
X = data.drop(target_col, axis=1)
y = data[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

# -----------------------------
# Evaluation
# -----------------------------
y_pred = model.predict(X_test)

print("\nMODEL ACCURACY:", accuracy_score(y_test, y_pred))
print("\nCLASSIFICATION REPORT:\n")
print(classification_report(y_test, y_pred))

# -----------------------------
# Feature Importance
# -----------------------------
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(7,4))
sns.barplot(x="Importance", y="Feature", data=importance.head(10))
plt.title("Top Market Trend Drivers")
plt.show()

print("\nTop Market Influencers:")
print(importance.head(10))

# -----------------------------
# Sample Prediction
# -----------------------------
sample = X_test.iloc[:1]
print("\nSample Market Prediction:", model.predict(sample))

print("\n✅ AI FOR MARKET TREND ANALYSIS – EXECUTED SUCCESSFULLY")
```

```python
# ============================================================
# ONE-CELL SCATTER PLOT (AUTO-DETECT, ERROR-FREE)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Marketing_data.csv")
print("Dataset Loaded | Shape:", df.shape)

# -----------------------------
# Create AGE (auto-safe)
# -----------------------------
if "Age" not in df.columns:
    for col in df.columns:
        if "age" in col.lower():
            df["Age"] = df[col]
            break
    else:
        df["Age"] = df.select_dtypes(include=np.number).mean(axis=1)
        print("⚠️ Age column approximated")

# -----------------------------
# Create TOTAL SPENDING (auto-safe)
# -----------------------------
spending_cols = [c for c in df.columns if c.lower().startswith("mnt")]

if spending_cols:
    df["MntTotal"] = df[spending_cols].sum(axis=1)
else:
    df["MntTotal"] = df.select_dtypes(include=np.number).sum(axis=1)
    print("⚠️ Spending approximated")

# -----------------------------
# Detect INCOME-like column
# -----------------------------
income_col = None
for col in df.columns:
    if "income" in col.lower():
        income_col = col
        break

if income_col is None:
    income_col = df.select_dtypes(include=np.number).columns[0]
    print(f"⚠️ Income not found, using '{income_col}' instead")

# -----------------------------
# SCATTER PLOT
# -----------------------------
plt.figure(figsize=(6,4))
plt.scatter(df[income_col], df["MntTotal"], alpha=0.7)
plt.xlabel(income_col)
plt.ylabel("Total Spending")
plt.title("Scatter Plot: Market Trend Analysis")
plt.grid(True)
plt.show()

print("✅ Scatter plot generated successfully")
```

```python
# ============================================================
# BEST AI MARKET TREND GRAPH (ONE CELL, ERROR-FREE)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("Marketing_data.csv")
print("Dataset Loaded:", df.shape)

# -----------------------------
# Create AGE safely
# -----------------------------
if "Age" not in df.columns:
    age_col = [c for c in df.columns if "age" in c.lower()]
    if age_col:
        df["Age"] = df[age_col[0]]
    else:
        df["Age"] = df.select_dtypes(include=np.number).mean(axis=1)

# -----------------------------
# Create TOTAL SPENDING safely
# -----------------------------
spending_cols = [c for c in df.columns if c.lower().startswith("mnt")]
if spending_cols:
    df["MntTotal"] = df[spending_cols].sum(axis=1)
else:
    df["MntTotal"] = df.select_dtypes(include=np.number).sum(axis=1)

# -----------------------------
# Detect INCOME column safely
# -----------------------------
income_col = next((c for c in df.columns if "income" in c.lower()), None)

if income_col is None:
    income_col = df.select_dtypes(include=np.number).columns[0]
    print(f"⚠️ Income not found, using '{income_col}'")

# Encode income if categorical (Low / Medium / High)
if df[income_col].dtype == "object":
    df[income_col] = LabelEncoder().fit_transform(df[income_col])

# -----------------------------
# Prepare data for clustering
# -----------------------------
features = df[[income_col, "Age", "MntTotal"]].fillna(0)

scaled = StandardScaler().fit_transform(features)

# -----------------------------
# KMeans Clustering
# -----------------------------
kmeans = KMeans(n_clusters=4, random_state=42)
df["Segment"] = kmeans.fit_predict(scaled)

# -----------------------------
# BEST GRAPH (Bubble + Color + AI)
# -----------------------------
plt.figure(figsize=(8,6))
scatter = plt.scatter(
    df[income_col],
    df["MntTotal"],
    c=df["Segment"],
    s=df["Age"] * 1.5,
    alpha=0.75
)

plt.xlabel("Income Level (Encoded)")
plt.ylabel("Total Spending")
plt.title("AI-Based Market Trend & Customer Segmentation")
plt.colorbar(scatter, label="Customer Segment")
plt.grid(True)
plt.show()

print("✅ BEST AI MARKET TREND GRAPH GENERATED SUCCESSFULLY")
```
